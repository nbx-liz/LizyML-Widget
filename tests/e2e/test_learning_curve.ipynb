```python
from sklearn.datasets import load_breast_cancer

from lizyml_widget import LizyWidget

data = load_breast_cancer(as_frame=True)
df = data.frame

w = LizyWidget()
w.load(df, target="target")

# Set 3 custom metrics including precision_at_k with k=20
w.set_config(
    {
        "model": {
            "name": "lgbm",
            "params": {
                "n_estimators": 50,
                "learning_rate": 0.1,
                "metric": ["auc", "binary_logloss", "precision_at_k"],
                "_precision_at_k_k": 20,
            },
        },
        "training": {
            "seed": 42,
            "early_stopping": {"enabled": True, "rounds": 10},
        },
    }
)

w.fit()
w
```
